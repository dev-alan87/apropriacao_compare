BIBLIOTECAS (imports)

```python
import nbimporter
import pandas as pd
import locale
from datetime import datetime
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

# desenvolvido por alan.alves@unesp.br
import compare_reports as compare # type: ignore
```

FUNÇÕES (desenvolvidas)

```python
def get_columns() -> list:
    columns = ['', 'ferias', '13_ferias', 'total_ferias', '13_salario', 'total']
    return columns

def format_value(value):
    try:
        if type(value) is str:
            value = value.strip()
        value = float(value)
        return locale.format_string('%1.2f', value, grouping=True)
    except ValueError:
        return value

def format_dataframe_to_excel(origin:str, df:pd.DataFrame) -> pd.DataFrame:
    adapted = pd.DataFrame(columns=get_columns())
    adapted.loc[0] = [origin.upper(), '', '', '', '', '']
    df.columns = get_columns()
    df.reset_index(drop=True, inplace=True)
    adapted = pd.concat([adapted, df.apply(lambda column: column.map(format_value))], ignore_index=True)
    return adapted
```

EXECUÇÃO PRINCIPAL (carrega e compara dados)

```python
cobol_file_path = "RHPP05LA.PDF"
java_file_path  = "RHPP05LA - GERAL.pdf"

results = compare.get_comparative_report(cobol_file_path, java_file_path)
```

EXECUÇÃO PRINCIPAL (impressão para excel)

```python
excel_file_name = datetime.now().strftime('%Y%m%d%H%M')+'_comparativo_RHPP05LA.xlsx'
excel_writer = pd.ExcelWriter(excel_file_name, engine='xlsxwriter')

for chave, valores in results.items():
    for tipo_funcionario, tabelas in valores.items():
        for tipo_tabela, origens in tabelas.items():
            sheet_name = f'{chave}_{tipo_funcionario}_{tipo_tabela}'
            sheet = pd.DataFrame(columns=get_columns())
            for origem, df in origens.items():
                sheet = pd.concat([sheet, format_dataframe_to_excel(origem, df)], ignore_index=True)
            sheet.to_excel(excel_writer, sheet_name=sheet_name, index=False)

excel_writer.close()
```
